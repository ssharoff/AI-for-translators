```python
import requests
```

```python
import re
```

```python
corpus_count = 0
dictionary = {}
text = requests.get("https://ssharoff.github.io/modl5007/parrot.txt").text
```

```python
print(len(text))
text[:100]
```

```python
text = text.lower()
text[:100]
```

```python
for sentence in text.split("."):
    for word in sentence.split(" "):
        corpus_count += 1
        word = re.sub("[^a-z-]+","",word)
        if word in dictionary:
          dictionary[word] += 1
        else:
          dictionary[word] = 1
```

```python
output_count = 0
for word in dictionary:
    ipm = dictionary[word] / (corpus_count / 1000000 )
    if ipm >= 10000:
        print(word, ipm)
        output_count += 1
print('Most frequent words count: ', output_count)
```
