```python
# prompt: convert a web page from a url into text using trafilatura library

# Fetch and extract text from a URL
url_fr = "https://fr.wikipedia.org/wiki/Jus_gentium"
url_ru = "https://ru.wikipedia.org/wiki/Jus_gentium"
url_en = 'https://en.wikipedia.org/wiki/Jus_gentium'
url_zh = 'https://zh.wikipedia.org/wiki/%E4%B8%87%E6%B0%91%E6%B3%95'
url_ja = 'https://ja.wikipedia.org/wiki/%E4%B8%87%E6%B0%91%E6%B3%95'
downloaded = trafilatura.fetch_url(url_zh)
text = trafilatura.extract(downloaded)
```

```python
from transformers import pipeline, AutoTokenizer, AutoModelForTokenClassification
```

```python
# modelnameTE = "tthhanh/xlm-ate-nobi-en"
modelnameTE = "tthhanh/xlm-ate-nobi-mul"
tokenizerTE = AutoTokenizer.from_pretrained(modelnameTE)
modelTE = AutoModelForTokenClassification.from_pretrained(modelnameTE)
classifierTE = pipeline("ner", model=modelTE, tokenizer=tokenizerTE)
```

```python
pred=classifierTE(text)
```

```python
import collections
dict = collections.defaultdict(int)
current_term='' # empty at the start
for word in pred:
  if len(word)>0:
    entity = word['entity']
    if entity=='LABEL_0':
      dict[current_term] += 1
      current_term = ''
    else:
      current_term += ' ' + word['word']
```
