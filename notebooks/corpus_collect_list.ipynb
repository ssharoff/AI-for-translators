```python
def process_url(url_name,corpus_dir,file_id):
    print("Started processing " + url_name)
    downloaded = trafilatura.fetch_url(url_name)
    plain_text = trafilatura.extract(downloaded)
    print("Saving it as " + file_id)
    out_file = open(corpus_dir + file_id + ".txt",mode="w")
    print(url_name, file=out_file)
    print(plain_text, file=out_file)
    out_file.close()
```

```python
corpus_dir = gdrive + "MyDrive/" + "corpus/law/english/"
topic_list = ["International law","Human rights","Ius gentium"]
top_url="https://en.wikipedia.org/wiki/"
for (i,topic) in enumerate(topic_list):
    url = top_url + topic
    process_url(url,corpus_dir,str(i))
```
