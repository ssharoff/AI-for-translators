```python
from transformers import pipeline, AutoTokenizer, AutoModelForSequenceClassification, AutoModelForTokenClassification
```

```python
modelnameEmotions = "bhadresh-savani/distilbert-base-uncased-emotion"
tokenizerEmotions = AutoTokenizer.from_pretrained(modelnameEmotions)
modelEmotions = AutoModelForSequenceClassification.from_pretrained(modelnameEmotions)
classifierEmotions = pipeline("text-classification", model=modelEmotions, tokenizer=tokenizerEmotions)
```

```python
modelnameGenres = "ssharoff/genres"
tokenizerGenres = AutoTokenizer.from_pretrained(modelnameGenres)
modelGenres = AutoModelForSequenceClassification.from_pretrained(modelnameGenres)
classifierGenres = pipeline("text-classification", model=modelGenres, tokenizer=tokenizerGenres)
```

```python
text=open(fname).read()
print(classifierGenres(text,top_k=2, truncation=True))
print(classifierEmotions(text,top_k=2, truncation=True))
```

```python
modelnameNER = "elastic/distilbert-base-uncased-finetuned-conll03-english"
tokenizerNER = AutoTokenizer.from_pretrained(modelnameNER)
modelNER = AutoModelForTokenClassification.from_pretrained(modelnameNER)
classifierNER = pipeline("ner", model=modelNER, tokenizer=tokenizerNER)
```

```python
modelnameTE = "tthhanh/xlm-ate-nobi-en"
tokenizerTE = AutoTokenizer.from_pretrained(modelnameTE)
modelTE = AutoModelForTokenClassification.from_pretrained(modelnameTE)
classifierTE = pipeline("ner", model=modelTE, tokenizer=tokenizerTE)
```

```python
def find_types_file(fname,model_type):
  text=open(fname).readlines()
  pred=model_type(text)
  dict = collections.defaultdict(int)
  for word in pred:
    if len(word)>0:
      curentity = word[0]['entity'][2:]
      curword = ''
      for subw in word:
        w = subw['word']
        if w.startswith("##"):
          curword += w[2:]
        else:
          curword += " " + w
      dict[curword+'/'+curentity] += 1
    return dict
```
